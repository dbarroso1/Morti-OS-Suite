```python
# Plots a radar chart.
from math import pi
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

### Set data

```python
# Categories
cat = ['anger', 'fear', 'happiness', 'sadness','surprise']

# Values
# Assuming that Morti will give back Floats between 1 & 0, I Multiply the values by 100
values = np.random.uniform(1,0,6) * 100

# Get Length of the List
N = len(cat)

# Use length to do somthing...
x_as = [n / float(N) * 2 * pi for n in range(N)]
```

### Formatting Data

```python
# Because our chart will be circular we need to append a copy of the 
# first value of each list at the end of each list with data
values += values[:1]
x_as += x_as[:1]
```

### Creating the Graph

```python
# Set color of axes
plt.rc('axes', linewidth=0.5, edgecolor="#888888")

# Create polar plot
ax = plt.subplot(111, polar=True)

# Set clockwise rotation. That is:
ax.set_theta_offset(pi / 2)
ax.set_theta_direction(-1)

# Set position of y-labels
ax.set_rlabel_position(0)

# Set color and linestyle of grid
ax.xaxis.grid(True, color="#888888", linestyle='solid', linewidth=0.5)
ax.yaxis.grid(True, color="#888888", linestyle='solid', linewidth=0.5)

# Set number of radial axes and remove labels
plt.xticks(x_as[:-1], [])

# Set yticks
plt.yticks([20, 40, 60, 80, 100], ["0.2", "0.4", "0.6", "0.8", "1"])

# Fill area
ax.fill(x_as, values, 'red', alpha=0.3)

# Set axes limits
plt.ylim(0, 100)
plt.title("EQE")

# Plot data
ax.plot(x_as, values, linewidth=0, linestyle='solid', zorder=3)

# Draw ytick labels to make sure they fit properly
for i in range(N):
    angle_rad = i / float(N) * 2 * pi

    if angle_rad == 0:
        ha, distance_ax = "center", 10
    elif 0 < angle_rad < pi:
        ha, distance_ax = "left", 1
    elif angle_rad == pi:
        ha, distance_ax = "center", 1
    else:
        ha, distance_ax = "right", 1

    ax.text(angle_rad, 100 + distance_ax, cat[i], size=10, horizontalalignment=ha, verticalalignment="center")

```
